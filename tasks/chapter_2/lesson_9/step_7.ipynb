```python
import pandas as pd
```

```python
df = pd.read_csv('../../../resources/submissions_data_train.csv')
df.head()
```

```python
df[df.submission_status == 'wrong'].step_id.value_counts().sort_values()
```

```python
df.sort_values(['user_id', 'timestamp'], ascending=False)
```

```python
df_with_uniq_user = df.sort_values(['user_id', 'timestamp'], ascending=False).drop_duplicates('user_id')
df_with_uniq_user
```

```python
df_for_grouping = df_with_uniq_user[df_with_uniq_user.submission_status == 'wrong'].drop(['timestamp', 'submission_status'], axis=1)
df_for_grouping
```

```python
df_for_grouping.groupby('step_id').count().sort_values('user_id', ascending=False).head(1)
```
